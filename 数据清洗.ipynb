# 联立抽血表和给药表

```python
# --- 第一步：Lab 数据透视 (加入 specimen_id) ---
lab_pivot = lab_features.pivot_table(
    index=['subject_id', 'hadm_id', 'stay_id','charttime'], #'specimen_id',
    columns='itemid', 
    values='valuenum',
    #aggfunc='mean' # 如果同一个样本对同一个指标测了多次，取均值
).reset_index()

# 重命名列名
lab_pivot = lab_pivot.rename(columns={
    51009: 'vanco_level', 
    50912: 'creatinine', 
    51301: 'wbc', 
    51006: 'bun'
})
lab_pivot
```

```python
# --- 第二步：计算 Lab 的相对时间 (相对于 intime) ---
lab_pivot = pd.merge(lab_pivot, df_patient[['stay_id', 'intime']], on='stay_id', how='left')
lab_pivot['rel_time'] = (pd.to_datetime(lab_pivot['charttime']) - pd.to_datetime(lab_pivot['intime'])).dt.total_seconds() / 3600
lab_pivot
```

```python
lab_pivot_copy = lab_pivot
df_patient_input_copy = df_patient_input

vanco_patient_stays = df_patient_input_copy['stay_id'].unique()

# 1. 统一标注事件类型，方便后续 RL 识别 Observation 和 Action
lab_pivot_copy['event_type'] = 'lab'
lab_pivot_copy_filtered = lab_pivot_copy[lab_pivot_copy['stay_id'].isin(vanco_patient_stays)].copy()
df_patient_input_copy['event_type'] = 'medication'
df_patient_input_copy['rel_time'] = df_patient_input_copy['vanco_start_rel']

# 2. 纵向合并
# concat 会保留两表所有的列。
# 给药行在 'vanco_level', 'creatinine' 等列会显示 NaN；
# 化验行在 'totalamount_mg' 等列会显示 NaN。
df_combined = pd.concat([df_patient_input_copy, lab_pivot_copy_filtered], axis=0, ignore_index=True)

# 3. 按病人 ID 和相对时间排序，生成真实的临床轨迹
# 这样你就得到了每个病人：化验 -> 给药 -> 再化验 -> 再给药 的序列
df_combined = df_combined.sort_values(['stay_id', 'rel_time']).reset_index(drop=True)
static_cols = ['subject_id', 'hadm_id', 'gender', 'anchor_age', 'patientweight', 'intime', 'outtime']
df_combined[static_cols] = df_combined.groupby('stay_id')[static_cols].ffill()
df_combined[static_cols] = df_combined.groupby('stay_id')[static_cols].bfill()

df_combined
```

# 重排列名

```python
# 1. 定义你希望看到的列顺序逻辑
# 核心 ID 和时间
id_cols = ['subject_id', 'hadm_id', 'stay_id', 'rel_time', 'event_type']

# 给药相关 (Action)
action_cols = ['itemid', 'totalamount_mg', 'starttime', 'vanco_start_rel', 'vanco_end_rel']

# 化验指标 (State/Observation)
observation_cols = ['vanco_level', 'creatinine', 'wbc', 'bun', 'charttime', 'specimen_id']

# 静态特征
static_cols = ['gender', 'anchor_age', 'patientweight', 'intime', 'outtime']

# 2. 合并所有列名，并剔除可能不存在的列（防止报错）
all_ordered_cols = id_cols + action_cols + observation_cols + static_cols
existing_cols = [c for c in all_ordered_cols if c in df_combined.columns]

# 3. 补充那些没在列表里但在原表里的列（以防遗漏）
remaining_cols = [c for c in df_combined.columns if c not in existing_cols]

# 4. 执行重排
df_combined = df_combined[existing_cols + remaining_cols]

# 5. 排序确保时间线正确
df_combined = df_combined.sort_values(['stay_id', 'rel_time']).reset_index(drop=True)
df_combined
```

# 最后联立 Chartevents

```python
# 1. 确保时间格式
chart_features['vitaltime'] = pd.to_datetime(chart_features['charttime'])

# 2. 计算相对时间 rel_time
chart_features = pd.merge(chart_features, df_patient[['stay_id', 'intime']], on='stay_id', how='inner')
chart_features['rel_time'] = (chart_features['vitaltime'] - pd.to_datetime(chart_features['intime'])).dt.total_seconds() / 3600

# 3. 分别对数值和报警标志进行透视
# 透视数值 (valuenum)
vitals_values = chart_features.pivot_table(
    index=['subject_id', 'stay_id', 'rel_time', 'vitaltime'],
    columns='itemid',
    values='valuenum',
    aggfunc='mean'
).reset_index().rename(columns={223762: 'temperature', 220179: 'sbp', 220045: 'heart_rate'})

# 透视报警标志 (warning)
# 我们加上 _warn 后缀以区分
vitals_warns = chart_features.pivot_table(
    index=['subject_id', 'stay_id', 'rel_time', 'vitaltime'],
    columns='itemid',
    values='warning',
    aggfunc='max' # 如果同一时刻有多个记录，只要有一个报警就记为 1
).reset_index().rename(columns={223762: 'temp_warn', 220179: 'sbp_warn', 220045: 'hr_warn'})

# 4. 合并数值和报警标志列
chart_pivot = pd.merge(vitals_values, vitals_warns, on=['subject_id', 'stay_id', 'rel_time', 'vitaltime'])
chart_pivot['event_type'] = 'vitals'

vanco_patient_stays = df_combined_copy['stay_id'].unique()
chart_pivot_filtered = chart_pivot[chart_pivot['stay_id'].isin(vanco_patient_stays)].copy()

chart_pivot_filtered
```

```python
chart_pivot_copy = chart_pivot_filtered
df_combined_copy = df_combined

df_combined_v3 = pd.concat([df_combined_copy, chart_pivot_copy], axis=0, ignore_index=True)
df_combined_v3 = df_combined_v3.sort_values(['stay_id', 'rel_time']).reset_index(drop=True)

# 3. 补全静态特征与基本信息
# 确保新加入的 vitals 行也拥有背景数据
static_cols = ['subject_id', 'hadm_id', 'gender', 'anchor_age', 'patientweight', 'intime']
df_combined_v3[static_cols] = df_combined_v3.groupby('stay_id')[static_cols].ffill()
df_combined_v3[static_cols] = df_combined_v3.groupby('stay_id')[static_cols].bfill()

df_combined_v3
```
